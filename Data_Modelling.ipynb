```python
# Import our dependencies
import pandas as pd
import matplotlib as plt
from sklearn.datasets import make_blobs
import sklearn as skl
```

```python
listing_df = pd.read_csv("../Resources/Housing_Market_Inventory_US.csv")
listing_df 
```

```python
listing_df["Year"] = listing_df["date"].str[0:4]
```

```python
listing_df["Month"] = listing_df["date"].str[8:10]
listing_df[['Year','Month']].dtypes
```

```python
get_year = lambda x: int(x.split("-")[0])
get_month = lambda x: int(x.split("-")[1])

pen = listing_df.copy()
pen['year_2'] = pen['date'].apply(get_year)
pen['month_2']=pen['date'].apply(get_month)
pen[['year_2','month_2']].dtypes

pen.head()
```

```python
pen.isna().sum()
```

```python
%%time
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

pen = listing_df.copy().sample(100)

pen = pen.dropna()

pen['year_2'] = pen['date'].apply(get_year)
pen['month_2']=pen['date'].apply(get_month)
pen[['year_2','month_2']].dtypes

pen = pen.drop(['date', 'state','Year','Month', 'city'],axis=1)

Y = pen['median_listing_price'].values
X = pen.drop('median_listing_price', axis=1).values

state = 42

x_train, x_test, y_train, y_test = train_test_split(X, Y, 
                                                    test_size=0.25, 
                                                    random_state= state)

model = RandomForestClassifier()

model.fit(x_train, y_train)

y_pred = model.predict(x_test)

# squared = false is for rmse,  squared = true then that is mse
mean_squared_error(y_test, y_pred, squared=False)
```
