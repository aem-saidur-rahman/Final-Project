```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime 
import numpy as np
```

```python
df = pd.read_csv("../Resources/Housing_Market_Inventory_US.csv")
df
```

```python
df.columns
```

```python
df.info()
```

```python
df.groupby(["postal_code", "zip_name"]).count()
```

```python
df["state"] = df["zip_name"].map(lambda x: x.split(',')[1].replace(" ", ""))
df["city"] = df["zip_name"].map(lambda x: x.split(',')[0].replace(" ", ""))
```

```python
places_columns = ["postal_code", "state", "city"]
places_df = df[places_columns].drop_duplicates()
listings_df = df.drop(columns = places_columns)
```

```python
places_df
```

```python
listings_df["listing_id"] = range(listings_df.shape[0])
```

```python
listings_df["postal_code"] = df["postal_code"]
```

```python
listings_df
```

```python
listings_df.columns
```

```python
listings_df.notnull().sum()
```

```python
listings_df.dropna(inplace = True)
```

```python
listings_df.info()
```

```python
# Convert the date column month_date_yyyymm into string first

listings_df['month_date_yyyymm'] = listings_df['month_date_yyyymm'].astype(str)
```

```python
# Insert two additional columns as Month and Year

listings_df['Year'] = listings_df['month_date_yyyymm'].str[0:4]
listings_df['Month'] = listings_df['month_date_yyyymm'].str[4:6]
```

```python
# Create column Date to concatenate the values from the Year and Month columns
listings_df["Date"] = pd.to_datetime(listings_df["Year"] + "/" + listings_df["Month"])
```

```python
# Change datatype to Datetime
listings_df['Date'] = pd.to_datetime(listings_df["Date"].dt.strftime('%Y/%m'))
```

```python
# Drop the month_date_yyyymm , Year and Mpnth columns
listings_df.drop(['month_date_yyyymm'], axis=1, inplace=True)
listings_df.drop(['Year'], axis=1, inplace=True)
listings_df.drop(['Month'], axis=1, inplace=True)
```

```python
# Rearrange the columns
listings_df=listings_df[['Date', 'postal_code', 'zip_name', 'median_listing_price', 'active_listing_count',
       'median_days_on_market', 'new_listing_count', 'price_increased_count',
       'price_reduced_count', 'pending_listing_count',
       'median_listing_price_per_square_foot', 'median_square_feet',
       'average_listing_price', 'total_listing_count', 'pending_ratio',
       'quality_flag', 'listing_id']]
```

```python
listings_df.head()
```

```python
listings_df.drop(['active_listing_count', 'new_listing_count', 'price_increased_count', 'price_reduced_count', 'pending_listing_count', 'pending_ratio',
       'quality_flag'], axis=1, inplace=True)
```

```python
listings_df.info()
```

```python
listings_df.head()
```

```python
# import the csv
listings_df.to_csv('../Resources/Cleaned/Cleaned_housing.csv', index=False)
```
