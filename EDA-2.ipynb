```python
df.groupby(["postal_code", "zip_name"]).count()
```

```python
df["state"] = df["zip_name"].map(lambda x: x.split(',')[1].replace(" ", ""))
df["city"] = df["zip_name"].map(lambda x: x.split(',')[0].replace(" ", ""))
```

```python
places_columns = ["postal_code", "state", "city"]
places_df = df[places_columns].drop_duplicates()
listings_df = df.drop(columns = places_columns)
```

```python
places_df
```

```python
listings_df["listing_id"] = range(listings_df.shape[0])
```

```python
listings_df["postal_code"] = df["postal_code"]
```

```python
listings_df
```

```python
listings_df.columns
```

```python
import sqlalchemy
from sqlalchemy import create_engine

path = "sqlite:///testo.sqlite"
conn = create_engine(path)
```

```python
listings_df.to_sql("listings", conn)
```

```python
places_df.to_sql("places", conn)
```
